## title

```python
%%manim -v WARNING -ql Title  
#-qk full hd 4k

from manim import *

class Title(Scene):
    def construct(self):
        texto = Text("Árboles de Decisión", font_size=75)
        
        self.play(DrawBorderThenFill(texto))
        self.wait(1)
        self.play(FadeOut(texto), run_time = 1)
```

## Quote 

```python
%%manim -v WARNING -ql Quote

class Quote(Scene):
    def construct(self):
        quote = Text("El árbol de decisión es una visualización \ndel proceso de pensamiento", font_size = 54)
        author = Text("-Garry Kasparov",font_size=54, color = YELLOW)

        quote.shift(UP * 2.5)
        author.shift(UP*0.4, RIGHT * 2.7)
        self.play(FadeIn(quote), run_time = 0.5)
        self.wait(2)
        self.play(Write(author))
        self.wait(1)
        self.play(FadeOut(quote,author), run_time = 0.5)
```

```python
%%manim -v WARNING -ql TextOnlyTree  
#-qk full hd 4k

class TextOnlyTree(Scene):
    def construct(self):
        # Nodo raíz
        root = Text("¿Está lloviendo?", font_size=36, color=YELLOW)
        root.move_to(UP * 2)
        
        # Ramas y decisiones
        branch_yes = Text("Sí", font_size=28, color=GREEN)
        decision_yes = Text("Llevar paraguas", font_size=28, color=GREEN)
        
        branch_no = Text("No", font_size=28, color=RED)
        decision_no = Text("No llevar paraguas", font_size=28, color=RED)
        
        # Posicionar
        branch_yes.move_to(UP * 0.5 + LEFT * 2)
        decision_yes.move_to(DOWN * 0.5 + LEFT * 2)
        
        branch_no.move_to(UP * 0.5 + RIGHT * 2)
        decision_no.move_to(DOWN * 0.5 + RIGHT * 2)
        
        # Líneas simples
        line1 = Line(root.get_bottom() + LEFT * 0.5, branch_yes.get_top(), color=GREEN)
        line2 = Line(branch_yes.get_bottom(), decision_yes.get_top(), color=GREEN, stroke_width=2)
        
        line3 = Line(root.get_bottom() + RIGHT * 0.5, branch_no.get_top(), color=RED)
        line4 = Line(branch_no.get_bottom(), decision_no.get_top(), color=RED, stroke_width=2)
        
        # Mostrar TODO
        self.add(root, branch_yes, decision_yes, branch_no, decision_no,
                line1, line2, line3, line4)
        self.wait(3)
```
